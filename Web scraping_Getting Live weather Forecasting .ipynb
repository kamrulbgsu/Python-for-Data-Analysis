# All code togather

```python
## Importing necessary library
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import io

# Connecting with url and creating soup
url = "https://www.accuweather.com/en/us/bowling-green/43402/current-weather/330106"
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.content, 'html.parser')

# Extracting initial data from the soup
initial_data = soup.find_all("div", class_="current-weather-card card-module content-module non-ad")

# Getting weather data from the initial data
for data in initial_data:
    temperature = data.find("div", class_="display-temp").text
    time= data.find("p", class_="sub").text
    others = data.find("div", class_= "current-weather-details").text

# Preparing all data for creating pandas data frame as csv format
weather_details = ["Current weather information"+"\n", "Time "+"\n"+time+"\n","Temperature "+"\n", temperature+ "\n", others]
join_weather_details =' '.join(weather_details)
space_remove_data = join_weather_details.replace("\n\n", '')
weather_data = io.StringIO(space_remove_data)
weather_data_organized = pd.read_csv(weather_data, sep=",")

# Creating subdata for organizing a suitable table    
weather_data_sub1 = (weather_data_organized.iloc[0:20:2]).reset_index()
weather_data_sub2= (weather_data_organized.iloc[1:21:2]).reset_index()

# Concatinating sub tables agian to produce an organized table
Live_weather_update = (pd.concat([weather_data_sub1,weather_data_sub2], axis = 1)).drop(['index'], axis= 1).set_axis(["Weather Variables", "Value"], axis=1)

# Exporting data to Excel as csv 
Live_weather_update.to_csv("C:\\Users\\msi\\Desktop\\Live_weather_update.csv")

# Show live weather update table
Live_weather_update
```
